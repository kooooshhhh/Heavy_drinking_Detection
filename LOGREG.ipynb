```python
#turning to train- test splits
train_split = int(0.75 *len(X))
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]
len(X_train), len(X_test) , len(y_train) , len(y_test)
```

```python
#builing a model
#subclas nn.Module
class DrunkModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1=nn.Linear(in_features=3,out_features=32)
        self.layer_2=nn.Linear(in_features=32,out_features=16)
        self.layer_3=nn.Linear(in_features=16,out_features=1)
        self.relu=nn.ReLU()
    
    def forward(self,x):
        return self.layer_3(self.relu(self.layer_2(self.relu(self.layer_1(x)))))
    
    
drunk_model=DrunkModel()
drunk_model
```

```python
#accuracy function
#calc accuracy
def accuracy_fn(y_true,y_pred):
    correct=torch.eq(y_true,y_pred).sum().item()
    acc=(correct/len(y_pred)) * 100
    return acc
```

```python
#view first 5 outputs of forward pass on test data
drunk_model.eval()
with torch.inference_mode():
    y_logits=drunk_model(X_test)[:5]
y_logits
```

```python
len(y_logits)
```

```python
#use sigmoid activation funtion on our model logits to tun them into prediction probablities
y_pred_probs = torch.sigmoid(y_logits)
y_pred_probs
```

```python
torch.round(y_pred_probs)
```

```python
#find the predicted lables
y_preds = torch.round(y_pred_probs)

#in full
y_pred_labels=torch.round(torch.sigmoid(drunk_model(X_test)))[:5]

# check for equality
print(torch.eq(y_preds.squeeze(),y_pred_labels.squeeze()))

# Check the shapes of y_preds and y_pred_labels
print(y_preds.shape)
print(y_pred_labels.shape)
#get rid of extra dimension
y_preds.squeeze()
```

```python
#loss function
loss_fn= nn.BCEWithLogitsLoss()
# #optimizer
# optimizer=torch.optim.SGD(drunk_model.parameters(),
#                           lr=0.1)
import torch.optim as optim

weight_decay = 0.001
optimizer = optim.Adam(drunk_model.parameters(), 
                       lr=0.001,
                       weight_decay=weight_decay )
```

```python
len(y_logits), len(y_train)
```

```python
#training loop
torch.manual_seed(42)

epochs=5000
for epoch in range(epochs):
    drunk_model.train()
    y_logits=drunk_model(X_train).squeeze(1)
    y_pred= torch.round(torch.sigmoid(y_logits))
    
    loss=loss_fn(y_logits,y_train)
    acc= accuracy_fn(y_true=y_train,y_pred=y_pred)
    
    optimizer.zero_grad()
    
    loss.backward()
    
    optimizer.step()
    
    drunk_model.eval()
    with torch.inference_mode():
        test_logits = drunk_model(X_test).squeeze(1)
        test_pred = torch.round(torch.sigmoid(test_logits))
        
        test_loss= loss_fn(test_logits,y_test)
        test_acc= accuracy_fn(y_true=y_test,y_pred=test_pred)
        
    if epoch%1000==0:
        print(f"Epoch: {epoch} | Train Loss: {loss: .5f},  Train Acc:{acc:.2f}% | Test loss: {test_loss:.5f}, Test acc: {test_acc:.2f}%")
        
```
